# Detecção de Faces com a biblioteca Face-Recognition

```python
# Para instalar a biblioteca face-recognition, execute o comando a seguir:
#!pip install face-recognition

import PIL.Image
import PIL.ImageDraw
import face_recognition
```

```python
imagem = face_recognition.load_image_file('imagens/faces.jpg')
```

```python
faces_encontradas = face_recognition.face_locations(imagem)

number_of_faces = len(faces_encontradas)
print("Encontramos {} face(s) nessa imagem.".format(number_of_faces))
```

```python
pil_image = PIL.Image.fromarray(imagem)
for face_encontrada in faces_encontradas:
    top, left, bottom, right = face_encontrada
    draw = PIL.ImageDraw.Draw(pil_image)
    draw.rectangle([left, top, right, bottom], outline="yellow", width=15)
```
