```python
from skimage import morphology, filters, io, color, feature
```

# Filtro Gaussiano

```python
img = io.imread('imagens/DKC2.jpg')
io.imshow(img)
io.show()
```

```python
gaussian_img  = filters.gaussian(img)
io.imshow(gaussian_img)
io.show()
```

# Filtro Eroso

```python
img = io.imread('imagens/DKC2.jpg')
io.imshow(img)
io.show()
```

```python
eroded_img  = morphology.erosion(img)
io.imshow(eroded_img)
io.show()
```

# Filtro Dilatao

```python
img = io.imread('imagens/DKC2.jpg')
io.imshow(img)
io.show()
```

```python
dilation_img  = morphology.dilation(img)
io.imshow(dilation_img)
io.show()
```

# Filtro Sobel


```python
img = io.imread('imagens/DKC2.jpg')
io.imshow(img)
io.show()
```

```python
img = color.rgb2gray(img)
sobel_img  = filters.sobel(img)
io.imshow(sobel_img)
io.show()
```

# Filtro Canny

```python
img = io.imread('imagens/DKC2.jpg')
io.imshow(img)
io.show()
```

```python
img = color.rgb2gray(img)
canny_img  = feature.canny(img,3)
io.imshow(canny_img)
io.show()
```

# Filtro Laplace

```python
img = io.imread('imagens/DKC2.jpg')
io.imshow(img)
io.show()
```

```python
img = color.rgb2gray(img)
sobel_img  = filters.laplace(img)
io.imshow(sobel_img)
io.show()
```
